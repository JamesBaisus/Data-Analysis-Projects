### Problem statement

* Evaluate the most proffeficent course among the dataset per Academic Year
* Find out which course is the most efficient among those courses per gender preference
* Prep the data for visualization

```python
import pandas as pd
import os
```

```python
cwd = os.getcwd()
file_path = os.path.join(cwd, 'appendix.csv')
dataset = pd.read_csv(file_path)
dataset.head()
```

#### Figuring out the col names of the dataset

```python
colum_names = dataset.columns
colum_names
```

#### Solving the first problem statement by figuraing out each courses with their respective academic years

```python
dataset['Year'].value_counts()
```

```python
year1 = dataset[dataset['Year'] == 1]
year2 = dataset[dataset['Year'] == 2]
year3 = dataset[dataset['Year'] == 3]
year4 = dataset[dataset['Year'] == 4]

year1_sorted_participants_data = year1.sort_values(by='Participants (Course Content Accessed)', ascending=False)
top_5_courses_freshman = year1_sorted_participants_data.head()

year2_sorted_participants_data = year2.sort_values(by='Participants (Course Content Accessed)', ascending=False)
top_5_courses_junior = year2_sorted_participants_data.head()

year3_sorted_participants_data = year3.sort_values(by='Participants (Course Content Accessed)', ascending=False)
top_5_courses_sophomore = year2_sorted_participants_data.head()

year4_sorted_participants_data = year4.sort_values(by='Participants (Course Content Accessed)', ascending=False)
top_5_courses_senior = year2_sorted_participants_data.head()



top_5_courses_freshman
```

```python
import matplotlib.pyplot as plt


import matplotlib.pyplot as plt

plt.figure(figsize=(9,7))


plt.subplot(2, 2, 1)
plt.bar(top_5_courses_freshman['Course Number'], top_5_courses_freshman['Participants (Course Content Accessed)'],
        width=0.5)
plt.xlabel('Course Code', labelpad=30)
plt.ylabel('Base Quantity of Participants', labelpad=30)
plt.title('Freshman')
plt.xticks(rotation=60)  

plt.subplot(2, 2, 2)
plt.bar(top_5_courses_junior['Course Number'], top_5_courses_junior['Participants (Course Content Accessed)'],
        width=0.5)
plt.xlabel('Course Code', labelpad=30)
plt.ylabel('Base Quantity of Participants', labelpad=30)
plt.title('Junior')
plt.xticks(rotation=60)  

plt.subplot(2, 2, 3)
plt.bar(top_5_courses_sophomore['Course Number'], top_5_courses_sophomore['Participants (Course Content Accessed)'],
        width=0.5)
plt.xlabel('Course Code', labelpad=30)
plt.ylabel('Base Quantity of Participants', labelpad=30)
plt.title('Sophomore')
plt.xticks(rotation=60)  

plt.subplot(2, 2, 4)
plt.bar(top_5_courses_senior['Course Number'], top_5_courses_senior['Participants (Course Content Accessed)'],
        width=0.5)
plt.xlabel('Course Code', labelpad=30)
plt.ylabel('Base Quantity of Participants', labelpad=30)
plt.title('Senior')
plt.xticks(rotation=60)  

plt.subplots_adjust(left=0.1, right=0.2, top=0.9, bottom=0.1, wspace=0.3, hspace=0.4)

plt.tight_layout()  
plt.show()



```
